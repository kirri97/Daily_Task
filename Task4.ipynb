5. Write a program that takes a sentence = 'Give papa a cup of proper in a copper
coffee cup' as input and add the word "coffee" after proper in the sentance.

```python
sentence = 'Give papa a cup of proper in a copper coffee cup'
insert = sentence.find("proper") + len("proper") + 1
sentence = sentence[:insert] + "coffee " + sentence[insert:]
print(sentence)
```
